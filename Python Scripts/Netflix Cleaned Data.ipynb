```python
df["date_added"] = pd.to_datetime(df["date_added"].str.strip(), errors="coerce") #changing datetype format
```

```python
df.dtypes
```

```python
df = df.apply(lambda col: col.str.strip() if col.dtype == "object" else col) #applying strip for all columns
```

```python
df
```

```python
space_counts = df.apply(lambda col: col.str.count(r'^\s+|\s+$').sum() if col.dtype == "object" else 0) #checking if any bogus column exists
print(space_counts)
```
